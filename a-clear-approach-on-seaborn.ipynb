# Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

This code tells us that the "dots" dataset is being loaded using the seaborn library (sns). The seaborn library is a data visualization library in Python that is built on top of matplotlib. It is commonly used to create attractive statistical graphics.

The `load_dataset` function is a function provided by seaborn that allows you to load one of the example datasets that are included with the library. The dataset is loaded as a Pandas DataFrame, which is a two-dimensional size-mutable tabular data structure with rows and columns that can hold any type of data.

The "dots" dataset is one of the example datasets provided with seaborn. It contains data on the reactions of mice to different combinations of stimuli. The dataset includes variables such as the time the mouse spent freezing (a measure of fear), the type of stimulus the mouse was exposed to, and the mouse's genotype.





```python
dots = sns.load_dataset("dots")
```

This code tells us that a line plot is being created using the seaborn library (sns) to visualize the relationship between two variables (x and y) in the "dots" dataset. The variables being plotted are "time" and "firing_rate", and the plot is being grouped by the "align" variable.

The `relplot` function is a function provided by seaborn that allows you to create scatter plots, line plots, or other kinds of plots that show the relationship between variables. In this case, the "kind" parameter is set to "line", which indicates that a line plot should be created.

The "dots" dataset is a Pandas DataFrame that was previously loaded using the `load_dataset` function provided by seaborn. The "time" and "firing_rate" variables are columns in this DataFrame, and the "align" variable is used to group the data by one of its categories.

The resulting plot will show the relationship between "time" and "firing_rate" for each level of the "align" variable. It will contain multiple lines, one for each level of "align", and the lines will be plotted in different colors or with different styles to distinguish between the groups.





```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align"
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    style='choice',hue='choice'
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    style='choice',hue='choice',size='coherence'
);
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    style='choice',hue='choice',size='coherence',
    facet_kws=dict(sharex=False),
);
```

This code tells us that a scatter plot with a fitted linear regression model is being created using the seaborn library (sns) to visualize the relationship between two variables (x and y) in the "dots" dataset. The variables being plotted are "time" and "firing_rate".

The `lmplot` function is a function provided by seaborn that allows you to create scatter plots with fitted linear regression models. It is similar to the `relplot` function, but it includes additional functionality for fitting and visualizing a linear regression model.

# Informative Distributional Summaries

This code tells us that a distribution plot is being created using the seaborn library (sns) to visualize the distribution of a single variable (x) in the "tips" dataset. The variable being plotted is "total_bill", and the plot is being grouped by the "time" variable.

The `displot` function is a function provided by seaborn that allows you to create plots that visualize the distribution of a single variable. It can create histograms, kernel density plots, or other types of plots that show the frequency or probability of different values in the data.

The "tips" dataset is a Pandas DataFrame that was previously loaded using the `load_dataset` function provided by seaborn. The "total_bill" and "time" variables are columns in this DataFrame.

The resulting plot will show the distribution of the "total_bill" variable for each level of the "time" variable. It will contain multiple plots, one for each level of "time", and the plots will be plotted in different colors or with different styles to distinguish between the groups. The plot may show the frequency or probability of different values of "total_bill" in the data, depending on the type of plot that was selected.

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips,x='total_bill',col='time');
```

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips,x='total_bill',col='time',kde=True);
#kde= kernel density estimate
```

```python
sns.displot(data=tips,kind='ecdf',x='total_bill',col='time');

#ECDF is an estimator of the Cumulative Distribution Function
```

# Classes & Functions for Making Complex Graphics

This code tells us that a PairGrid object is being created using the seaborn library (sns) for the "penguins" dataset. The PairGrid object is a multi-plot grid that can be used to visualize the pairwise relationships between variables in the dataset.

The `PairGrid` function is a function provided by seaborn that allows you to create a grid of subplots that can be used to visualize the pairwise relationships between variables in a dataset. It is similar to a scatter plot matrix, but it allows you to customize the types of plots that are shown in each subplot and to control the appearance of the plots.

The "penguins" dataset is a Pandas DataFrame that contains data on penguins. The "species" variable is a categorical variable that indicates the species of the penguin.

The `hue` parameter is used to specify that the plots in the PairGrid should be colored by the "species" variable. The corner parameter is used to specify that a plot should be added to the corner of the grid that shows the distribution of each variable separately.

The resulting PairGrid object will contain a grid of subplots that show the pairwise relationships between the variables in the "penguins" dataset. The plots will be colored by the "species" variable, and a plot showing the distribution of each variable separately will be included in the corner of the grid. You can then use various methods provided by the PairGrid object to customize the appearance of the plots or to plot additional data.





```python
g = sns.PairGrid(data = penguins,hue='species',corner=True)
```

```python
g = sns.PairGrid(data = penguins,hue='species',corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2");
g.map_lower(sns.scatterplot,marker='*');
```

```python
g = sns.PairGrid(data = penguins,hue='species',corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2");
g.map_lower(sns.scatterplot,marker='*')
g.map_diag(sns.histplot,element='step',kde=True);
```

```python
g = sns.PairGrid(data = penguins,hue='species',corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2");
g.map_lower(sns.scatterplot,marker='*')
g.map_diag(sns.histplot,element='step',kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6));
```

# Customizing Standardized Parameters

This code tells us that a scatter plot is being created using the seaborn library (sns) to visualize the relationship between two variables (x and y) in the "penguins" dataset, and to show the distribution of one additional variable (hue). The variables being plotted are "bill_length_mm", "bill_depth_mm", and "body_mass_g".

The `relplot` function is a function provided by seaborn that allows you to create scatter plots or other kinds of plots that show the relationship between variables. In this case, the "kind" parameter is not specified, so a scatter plot is created by default.

The "penguins" dataset is a Pandas DataFrame that contains data on penguins. The "bill_length_mm", "bill_depth_mm", and "body_mass_g" variables are columns in this DataFrame that contain measurements of the penguins' bill length, bill depth, and body mass, respectively.

The `set_theme` function is used to set the default style and font scale for all plots created with seaborn. The style parameter is set to "ticks" to show tick marks and labels on the axes, and the font_scale parameter is set to 1.25 to increase the size of the text in the plot.

The `hue` parameter is used to specify that the points in the scatter plot should be colored by the "body_mass_g" variable. The `palette` parameter is used to specify the color palette that should be used to map the values of the "body_mass_g" variable to colors. The marker parameter is used to specify the marker style that should be used to plot the points, and the `s` parameter is used to specify the size of the markers.

The resulting scatter plot will show the relationship between the "bill_length_mm" and "bill_depth_mm" variables, with the points colored by the "body_mass_g" variable. The plot will also show tick marks and labels on the axes, and the text in the plot will be larger than the default size.

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
```
